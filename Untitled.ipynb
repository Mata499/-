```python
import pandas as pd

# 读取 CSV 文件（假设第一列是待修改的日期列）
df = pd.read_csv("dataset.csv")

# 生成时间序列（按15分钟间隔，长度与数据行数一致）
start_time = "2018-06-30 00:00:00"
df.iloc[:, 0] = pd.date_range(start=start_time, periods=len(df), freq="15T")

# 保存修改后的 CSV 文件
df.to_csv("dataset1.csv", index=False)
```
